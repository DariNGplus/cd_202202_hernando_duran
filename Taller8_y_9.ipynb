```python
dfLogit = df.sample(frac=0.8)


dfTest = df.loc[~df.index.isin(dfLogit.index)]
dfTest.reset_index()
```

```python
x = dfLogit.iloc[:, [5,6,7]]
y = dfLogit.Survived

logit_model = sm.Logit(y,x)
result = logit_model.fit()
print(result.summary())
```

```python
import math

probOf = result.params.values
contFP = 0
contTP = 0
contFN = 0
contTN = 0

for row in range(len(dfTest)):
  value = 1/(1+ math.e**-(dfTest.iloc[row]['Age']*probOf[0] + dfTest.iloc[row]['SibSp']*probOf[1] + dfTest.iloc[row]['Parch']*probOf[2]))
  if (value < 0.5 and dfTest.iloc[row]['Survived'] == 0):
    contTN += 1
  elif (value < 0.5 and dfTest.iloc[row]['Survived'] == 1):
    contFN += 1
  elif (value >= 0.5 and dfTest.iloc[row]['Survived'] == 1):
    contTP += 1
  else:
    contFP += 1

precision = contTP /(contTP+contFP)

recall = contTP / (contTP+ contFN)
  

print("Precisión: " + to_str(precision))

print("Exhaustividad: " + to_str(recall))
```
