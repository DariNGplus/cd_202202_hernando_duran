```python
#1
def kde_quartic(d,h):
    dn=d/h
    P=(15/16)*(1-dn**2)**2
    return P
#2
def kde_epanechnikov(d,h):
    dn=d/h
    P=(3/4)*(1-dn**2)
    return P
#3
def kde_gaussian(d,h):
    dn=d/h
    P=math.e**(-(dn**2)/2)/math.sqrt(2*math.pi)
    return P
#4
def kde_bebesos(d,h):
    dn=d/h
    P=3/8*(3-5*dn**2)
    return P
#5
def kde_uniform():
  return 0.5
  
#6
def kde_triangular(d,h):
  dn = d/h
  return 1-abs(dn)


#SELECCIÓN DE FUNCIÓN KERNEL
def kde_selection(kernel_function, d, bw):
  kernel = 0
  if (kernel_function == 1):
    kernel = kde_quartic(d, bw)
  elif (kernel_function == 2):
    kernel = kde_epanechnikov(d,bw)
  elif (kernel_function == 3):
    kernel = kde_gaussian(d, bw)
  elif (kernel_function == 4):
    kernel = kde_bebesos(d, bw)
  elif (kernel_function == 5):
    kernel = kde_uniform()
  elif (kernel_function == 6):
    kernel = kde_triangular(d, bw)
  return kernel
      
```

```python
import numpy

bw = 100
res = 10

kernel_function = 1

xmin = df['x'].min() - bw
xmax = df['x'].max() + bw

ymin = df['y'].min() - bw
ymax = df['y'].max() + bw

rows = int(math.ceil(ymax-ymin)/res)

cols = int(math.ceil(xmax-xmin)/res)

map = numpy.zeros((rows, cols))


for i in range(df.shape[0]):
  curRow = int(math.floor((df['y'].iloc[i] - ymin)/res))
  curCol = int(math.floor((df['x'].iloc[i] - xmin)/res))
  ##print( str(curRow)+ " , " +str(curCol))
  map[curRow, curCol] += 1
```

```python
# IMPORTS PARA IMPRIMIR HEATMAP
import seaborn as sns
import matplotlib.pylab as plt
from matplotlib.pyplot import figure
```

```python
#RES = 10, BW= 100, KERNEL => QUARTIC

amap = numpy.zeros((rows, cols))
kernel =  0


for i in range(0, len(map)):
  for j in range(0, len(map[i])):
    if (map[i][j] != 0):
      for k in range(max(0, int(i-(bw/res))), min(rows, int(i+(bw/res)))):
        for l in range(max(0, int(j-(bw/res))), min(cols, int(j+(bw/res)))):
          
          d = math.sqrt((k-i)**2 + (l-j)**2)
          amap[k][l] += map[i][j]*kde_selection(1, d, bw)/(math.pi*bw**2)
      
figure(figsize=(12, 10), dpi=100)

ax = sns.heatmap(amap)

plt.show()
```

```python
import colorsys

def gen_color_palette(ri, gi, bi, rf, gf, bf, qt):
  hsv_init = colorsys.rgb_to_hsv(ri/255, gi/255, bi/255)
  hsv_fin = colorsys.rgb_to_hsv(rf/255, gf/255, bf/255)

  stepH = (hsv_fin[0]-hsv_init[0])/(qt-1)
  stepS = (hsv_fin[1]-hsv_init[1])/(qt-1)
  stepV = (hsv_fin[2]-hsv_init[2])/(qt-1)

  listRGB = []
  listRGB.append([ri/255, gi/255, bi/255])

  for i in range(qt-1):
    h = hsv_init[0] + stepH * (i+1)
    s = hsv_init[1] + stepS * (i+1)
    v = hsv_init[2] + stepV * (i+1)    
    rgb = colorsys.hsv_to_rgb(h, s, v)
    listRGB.append(rgb)


  return listRGB
```

```python
import matplotlib.image as mpimg


color = gen_color_palette(255, 255, 0, 255, 0, 0, 50)
map_img = mpimg.imread('mapaFloridablanca.png') 
amap = np.zeros((rows, cols))


for i in range(0, len(map)):
  for j in range(0, len(map[i])):
    if (map[i][j] != 0):
      for k in range(max(0, int(i-(bw/res))), min(rows, int(i+(bw/res)))):
        for l in range(max(0, int(j-(bw/res))), min(cols, int(j+(bw/res)))):
          d = math.sqrt((k-i)**2 + (l-j)**2)
          amap[k][l] += map[i][j] * kde_selection(1, d, bw)/(math.pi*bw**2)
figure(figsize=(12, 10), dpi=100)

mask = np.zeros((rows, cols))
for row in range(rows):
  for col in range(cols):
    if(amap[row][col] == 0):
      mask[row][col] = True
    else:
      mask[row][col] = False

hmax = sns.heatmap(amap, cmap=color, mask=mask)
hmax.collections[0].set_alpha(0.5)

plt.imshow(map_img, zorder=0, extent=[0.5, 450.0, 1.0, 468.0])

plt.show()
```
