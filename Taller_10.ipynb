```python
import numpy as np
import pandas as pd
import seaborn as sb
import os
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/master/titanic.csv")
df
```

```python
df = df.drop(columns =['Cabin', 'Ticket', 'Name', 'PassengerId'])
df
```

```python
df = df.dropna()
df
```

```python
df['Sex'] = np.where(df['Sex'] == 'female', 0, 1)
df
```

```python
df['Embarked'] = np.select([df['Embarked'] == 'C', df['Embarked'] == 'Q', df['Embarked'] == 'S'], [0, 1, 2])
df
```

```python
predict = df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
goal = df.Survived
```

```python
pred_train, pred_test, res_train, res_test = train_test_split(predict, goal, test_size=0.2)
```

```python
SVM_Model = SVC(gamma='auto', kernel='rbf', C=0.5)
SVM_Model.fit(pred_train, res_train)
```

```python
res_test
```

```python
print (f'Precisión: {SVM_Model.score(pred_test, res_test):.4f}')
```
